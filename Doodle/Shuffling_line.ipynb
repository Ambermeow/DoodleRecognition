```python
import pandas as pd
import numpy as np
import csv
import math
import os
import os.path
from sklearn.utils import shuffle
```

```python
for ifile in range(1,100):
    print(ifile)
    shuffled_filename="Data/Shuffled_train_data/train_"+str(ifile)+".csv"
    ff=pd.read_csv(shuffled_filename)
    del ff["Unnamed: 0"]
    ff=shuffle(ff)
    print(ff.head(10))
    with open(shuffled_filename, 'w') as f:
        ff.to_csv(f,header=True,index=False)
```
